**one hot encoder**

- one hot means at a time only one will **ON** and another will **OFF**

- **ON** represent with 1

- **OFF** represent with 0

- one hot encoder creates new columns

- for example case ctatus has two unique lables

- Number of new columns = Number of unique lables

- case_status has two unique lables

- Certified and Denied

- it will create two new columns

    - case_status_certified

    - case_status_Denied

- original column as prefix

|case_status|case_status_Certified|case_status_Denied
|-|-|-|
|Certified|1|0|
|Denied|0|1|
|Denied|0|1|
|Certified|1|0|

**Advantages**

- Genarally ML models excepects independence between input variables

- there is no relation between the variables

- because od one hot at a time only one will be on , other will off

- this indicates the new columns are independented each other

- this independent behaviour also called as **orthogonality**

- independent menas no relation

- no relation means perpendicular each other

- perpendicular menas 90 degress phase shift

- 90 degrees means orthogonality

**Disadvantage**

- it will create more columns

- morecolumns means more data

- more data means more complexity w.r.t time and infra

- this is calles as **curse of Dimensionality**
