**step-9: drop the column**

- in order to drop a column we need to use **drop** method

- all the methods based on your dataframe name similar as string names

- it requires mainly 3 arguments

    - 1. column name

    - 2. axis

         - axis=1 represent column

         - axis=0 represent rows

    - 3. inplace

         - once you drop the column , the dataframe affected

         - the modified dataframe wants to save in a same name or different name

         - if you want to keep at same name then **inplace=True**

```python
name=['Raj','Abhi','Chinna']
age=[20,21,22]
city=['hyd','mumbai','pune']
cols=['Names','Age','City']
idx=['A','B','C']
df=pd.DataFrame(zip(name,age,city),columns=cols,index=idx)
df.drop('City',axis=1,inplace=True)
df
```

```python
name=['Raj','Abhi','Chinna']
age=[20,21,22]
city=['hyd','mumbai','pune']
cols=['Names','Age','City']
idx=['A','B','C']
df=pd.DataFrame(zip(name,age,city),columns=cols,index=idx)
df.drop('B',axis=0,inplace=True)
df
```

**step-10: how to overwrite the existed column data**

- we already have a dataframe 

- now we want to replace all the values of a specific column with new values

- first create a list with new values 

- then update the column with new values , in the same way of how to create a new column

- df[new col]=data , to create a new column

- df[old city]= new data , to overwrite the old column

```python
name=['Raj','Abhi','Chinna']
age=[20,21,22]
city=['hyd','mumbai','pune']
cols=['Names','Age','City']
idx=['A','B','C']
df1=pd.DataFrame(zip(name,age,city),columns=cols,index=idx)
city=['chennai','bnglr','maharastra']
df1['City']=city
df1
```
